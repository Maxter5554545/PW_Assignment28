```python
# Q1. How can you create a Bokeh plot using Python code?
import logging
logging.basicConfig(filename="As28.log", level=logging.DEBUG)
#We can create Bokeh plot by-
try:
  import bokeh.io #Importing bokeh input & output system

  import bokeh.plotting #Importing bokeh.plotting for plotting the graphs

  from bokeh.plotting import figure, output_file, show

  from bokeh.sampledata.iris import flowers

  output_file('test.html')

  p = figure(title="TestFlower")

  p.xaxis.axis_label = "Petal length"

  p.yaxis.axis_label = "Petal width"

  p.circle(flowers['petal_length'], flowers['sepal_length'])

  show(p)
except ValueError as v:
  logging.error(v)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

try:
  print("Glyphs are nothing but basic visual representations of different shapes and patterns for the visualisation of graphs.")

  #We can add it just by adding different kinds of plots:

  output_file("Question_2")

  g= figure(title="plotting a plot")

  x = [1,2,3,4,5]
  y = [i**2 for i in x]

  g.xaxis.axis_label = "exponents"
  g.yaxis.axis_label = 'numbers'

  g.line(x,y)

  #scatter plot
  g.circle(x,y)

  #square plot
  g.square(x,y)

  #text plot

  g.text(x,y)
  show(g)
except ValueError as v:
  logging.error(v)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

#We can customize bokeh plot by using properties shown in the example:
try:
  output_file("q3.html")

  three = figure(title='three answer')

  x = [10, 19, 20, 21, 99, 88, 12, 22]
  y = [i for i in x if i%2==0]

  three.xaxis.axis_label = 'Numbers'

  three.yaxis.axis_label = "Numbers which can be divided by two"

  three.circle(x,y, color='red', size=20, legend_label = "number whose reaminder is zero")
  show(three)
except ValueError as v:
  logging.error(v)
```
